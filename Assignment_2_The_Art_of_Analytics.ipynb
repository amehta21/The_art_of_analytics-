Make a pie chart to analyse Teams (Either Team 1 or Team 2) with mumber of match played?
Label the Pie Chart.

```python


frequency = data['Team1'].value_counts()
frequency_df2 = pd.DataFrame({'Element': frequency.index, 'Frequency': frequency.values})

plt.pie(frequency_df2['Frequency'], labels = frequency_df2['Element'], autopct='%.0f%%')
plt.show()
```

Bar chart of Umpires (U can use Umpire 1 or Umpire 2, ur choice) and number of match they umpired?
Set figure size as (100,20)

```python
frequency = data['Umpire1'].value_counts()
frequency_df = pd.DataFrame({'Element': frequency.index, 'Frequency': frequency.values})

plt.figure(figsize=(100, 20))
plt.bar(frequency_df['Element'], frequency_df['Frequency'])
plt.xlabel('Element')
plt.ylabel('Frequency')
plt.title('Frequency of Column Elements')


plt.show()
```
