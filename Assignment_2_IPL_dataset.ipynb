Make a pie chart to analyse Teams (Either Team 1 or Team 2) with mumber of match played?

```python
value_counts = ipl_matches['Team1'].value_counts()

plt97.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')
plt97.title('Team v/s Matches played')
plt97.show()
```
